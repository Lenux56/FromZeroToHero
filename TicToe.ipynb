```python
def checkTic(inlist):
    succes = False
    xlist = ['X','X','X']
    olist = ['O','O','O']
    for i in range(2):
        if inlist[i] == xlist or inlist[i] == olist:
            print(f'The win is 1 {inlist[i]}')
            succes = True
            break
        if inlist[0][i] == inlist[1][i] == inlist[2][i] != ' ':
            print(f'The win is 2 {inlist[0][i]}')
            succes = True
            break
    if inlist[0][0] == inlist[1][1] == inlist[2][2] != ' ':
        print(f'The win is 3 {inlist[0][0]}')
        succes = True
    if inlist[0][2] == inlist[1][1] == inlist[2][0] != ' ':
        print(f'The win is 4 {inlist[0][2]}')
        succes = True
    return succes    

def endgame(inlist):
    for i in range(3):
        for j in range(3):
            if inlist[i][j] == ' ': 
                  return False
    return True

def printTic(tlist):
    for i in range(3):
        for j in range(3):
            print(tlist[i][j], end=' ')
        print()  
              
def state(sy, tlist):
    
    while True:
        first = int(input())-1
        if first == 100:
            print('Game stoped')
            return True 
        if first > 8 or first < 0:
            print('You write incorrect number. Try again, please') 
            continue
        if tlist[first // 3][first % 3] == ' ':
            tlist[first // 3][first % 3] = sy
            printTic(tlist)
            break
        elif endgame(tlist):
            print('Game over. All cells is not empty')
            return True
        else: 
            print('The cell is not empty. Try again, please')
    if checkTic(tlist):
        print('Congratulation!!!')
        return True
    
def TicToe():
    tlist = [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
    while True:
        print('First player is X. Please write a number from 1 to 9')
        if state('X', tlist):
            break
        print('Second player is O. Please write a number from 1 to 9')
        if state('O', tlist):
            break
        
    
    
```
